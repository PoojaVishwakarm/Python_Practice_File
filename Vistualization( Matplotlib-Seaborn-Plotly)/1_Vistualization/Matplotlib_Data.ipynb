```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X axis")
plt.ylabel("y axis")
```

## subplot in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

Object Oriented Way

```python
fig=plt.figure(figsize=(20,8),dpi=100)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.7,0.6])
axis2.plot(x,y)
```
