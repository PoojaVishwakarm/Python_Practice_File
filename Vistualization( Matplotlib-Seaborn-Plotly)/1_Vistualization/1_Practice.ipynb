```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sample diamond data
df = pd.read_csv('../0_datasets/diamonds.csv')
df
```

```python
# Display the first few rows of the dataset
df.head()
```

basic plots using matplotlib

```python
plt.plot(df['carat'],df['price'])
plt.title('carat vs price line plot')
plt.xlabel('carat')
plt.ylabel('price')
plt.show()
```

 Scatter plot

```python
plt.scatter(df['carat'],df['price'])
plt.title('carat vs price scatter plot')
plt.xlabel('carat')
plt.ylabel('price')
plt.show()
```

```python
# bar Plot
plt.bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
plt.title('Average price by cut')
plt.xlabel('cut')
plt.ylabel('Average price')
plt.show()
```

Intermediate Plots using Matplotlib

```python
#histogram
plt.hist(df['price'],bins=10,edgecolor='black')
plt.title('Distribution of pricess')
plt.xlabel('Price')
plt.ylabel('Frequancy')
plt.show()
```

```python
# Box Plot
plt.boxplot(df['carat'],vert=False) # horizontal box plot
plt.title('box plot of carat')
plt.xlabel('carat')
plt.show()
```

```python
cut_count=df['cut'].value_counts()
cut_count
```

```python
# pie Chart
cut_count=df['cut'].value_counts()
plt.pie(cut_count,labels=cut_count.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of cut Quality')
plt.show()
```

Advanced Plots  using matplotlib


```python
correlation=df[['carat','depth','table','price','x','y','z']].corr()
correlation
```

```python
# HeatMap of correclation
correlation=df[['carat','depth','table','price','x','y','z']].corr()
plt.imshow(correlation,cmap='coolwarm',interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation)),correlation.columns,rotation=90)
plt.yticks(range(len(correlation)),correlation.columns)
plt.title('Heatmap of correlation Matrix')
plt.show()
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```

```python
# subplots
fig,axes=plt.subplots(1,2,figsize=(12,5))
axes[0].hist(df['carat'],bins=10,edgecolor='black')
axes[0].set_title('Histogram of carat')
axes[1].hist(df['price'],bins=10,edgecolor='black')
axes[1].set_title('Histogram of Price')
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(12,5))
axes[0,0].hist(df['carat'],bins=10,edgecolor='black')
axes[0,0].set_title('Histogram of carat')
axes[0,1].hist(df['price'],bins=10,edgecolor='black')
axes[0,1].set_title('Histogram of Price')
axes[1,0].scatter(df['carat'],df['price'])
axes[1,0].set_title('carat vs price scatter plot')
axes[1,0].set_xlabel('carat')
axes[1,0].set_ylabel('price')
axes[1,1].bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
axes[1,1].set_title('Average price by cut')
axes[1,1].set_xlabel('cut')
plt.show()

plt.show()
```

Enhancing plots with seaborn for reports -ready Vistualoization

```python
# setting seaborn style
sns.set_style('whitegrid')
sns.set_context('talk')

sns.scatterplot(x='carat',y='price',data=df)
plt.title('Scatter plot with whitegrid style and notebook context')
plt.show()
```

```python
# using a dark backgroud style
sns.set_style('dark')
sns.lineplot(x='carat',y='price',data=df)
plt.title("Line plot with Dark Background Style")
plt.show()
```

```python
# Customizing the font scale
sns.set_context('talk',font_scale=1.5)
sns.boxplot(x='cut',y='price',data=df)
plt.title('box plot with talk context and larger font scale')
plt.show()
```

```python
# Using a predefined color palette
figure = plt.figure(figsize=(12, 3))
sns.set_context('notebook')
sns.set_palette('deep') # Options: deep, muted, bright, pastel, dark, colorblind
sns.violinplot(x='color', y='price', data=df, hue='cut')
plt.title('Violin Plot with Pastel Color Palette')
plt.show()
```

```python
# Creating and using a custom color palette
custom_palette = sns.color_palette(['#FF5733', '#33FFCE', '#335BFF', '#FF33A1', '#9B33FF'])
sns.set_palette(custom_palette)
sns.countplot(x='cut', data=df, hue='color')
plt.title('Count Plot with Custom Color Palette')
plt.show()
```

Adjusting Plot Elements for Better Presentation

```python
# Adding gridlines to the background
sns.set_style('whitegrid')
sns.scatterplot(x='carat', y='price', data=df)
plt.grid(True, linestyle='--', alpha=0.5)  # Adding dashed gridlines
plt.title('Scatter Plot with Gridlines')
plt.show()
```

```python
# Removing spines from the plot for a cleaner look
sns.set_style('white')
ax = sns.histplot(df['price'], bins=10, edgecolor='black')
sns.despine()  # Remove the top and right spines
plt.title('Histogram with Spines Removed')
plt.show()
```

Adding Annotations and Legends

```python
# Adding annotations to a scatter plot
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Scatter Plot with Annotations')
for i in range(len(df)):
    plt.annotate(f"${df['price'][i]}", (df['carat'][i], df['price'][i]), fontsize=8, alpha=0.7)
plt.show()
```

```python
# Customizing the legend
figure = plt.figure(figsize=(12, 3))
sns.boxplot(x='cut', y='price', data=df)
plt.legend(title='Cut Types', labels=df['cut'].unique(), loc='lower right', frameon=True)
plt.title('Box Plot with Custom Legend')
plt.show()
```

```python
# Customizing the legend
figure = plt.figure(figsize=(12, 3))
sns.boxplot(x='cut', y='price', data=df)
plt.legend(title='Cut Types', labels=df['cut'].unique(), loc='upper left', bbox_to_anchor=(1, 1), frameon=True)
plt.title('Box Plot with Custom Legend')
plt.show()
```


Using FacetGrid for Multi-Panel Plots

```python
# FacetGrid for multiple histograms based on cut
g = sns.FacetGrid(df, col='cut', col_wrap=3, height=4)
g.map(plt.hist, 'price', bins=10, edgecolor='black')
g.set_titles("{col_name} Cut")
plt.show()
```

```python
fig = plt.figure(figsize=(12, 6))
# Using FacetGrid with scatter plots and hue
g = sns.FacetGrid(df, hue='color', height=5)
g.map(plt.scatter, 'carat', 'price').add_legend()
plt.title('FacetGrid Scatter Plot with Hue')
plt.show()
```

Fine-Tuning Axis Labels and Titles

```python
# Rotating x-axis labels for better readability
sns.barplot(x='color', y='price', data=df)
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.title('Bar Plot with Rotated X-axis Labels')
plt.show()
```

```python

# Adjusting the y-axis scale
sns.histplot(df['price'], bins=10, edgecolor='black')
# plt.yscale('log')  # Use logarithmic scale for y-axis
plt.title('Histogram with Logarithmic Y-axis Scale')
plt.show()
```

```python

# Adjusting the y-axis scale
sns.histplot(df['price'], bins=10, edgecolor='black')
plt.yscale('log')  # Use logarithmic scale for y-axis
plt.title('Histogram with Logarithmic Y-axis Scale')
plt.show()
```

Using Seaborn Themes for Final Touches

```python
# Applying the 'ticks' theme
figure = plt.figure(figsize=(12, 3))
sns.set_theme(style='ticks', palette='muted', context='notebook', font_scale=1)
sns.lineplot(x='carat', y='price', data=df)
sns.despine()  # Remove spines for a cleaner look
plt.title('Line Plot with Ticks Theme and Muted Palette')
plt.show()
```
